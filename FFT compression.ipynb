```python
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt 

img = Image.open(r".\o1a3g3ra.bmp")

img = np.array(img)

img = np.mean(img,axis=2)
plt.imshow(img, cmap='gray')
fft = np.fft.fft2(a=img)
```

```python
print(fft)
```

```python
plt.imshow(fft.real, cmap='gray')
```

```python
imgsort = np.sort(np.abs(fft.reshape(-1)))

for keep in (0.1, 0.05, 0.01, 0.002):
    plt.figure()
    thresh = imgsort[int(np.floor((1-keep) * len(imgsort)))]
    ind = np.abs(fft)>thresh
    Atlow = fft * ind
    Alow = np.fft.ifft2(Atlow).real
    plt.imshow(Alow, cmap='gray')
    plt.axis('off')
    plt.title(f'Compression with keep={keep}')
```
