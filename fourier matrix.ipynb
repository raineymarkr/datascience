```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [15, 15]
plt.rcParams.update({'font.size': 18})

n = 256
w = np.exp(-1j * 2 * np.pi / n)

DFT = np.ndarray((n,n))
print(DFT.shape)
#slow
for i in range(n):
    for k in range(n):
        DFT[i,k] = w ** (i * k)
DFT = np.real(DFT)

plt.imshow(DFT)
```

```python
n = 2048
w = np.exp(-1j * 2 * np.pi / n)

DFT = np.ndarray((n,n))

#fast
J,K = np.meshgrid(np.arange(n), np.arange(n))
DFT = np.power(w, J*K)
DFT = np.real(DFT)

plt.imshow(DFT)
```
