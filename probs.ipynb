```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np

A = imread(r"C:\Users\mark.rainey\Downloads\data science\DATA\dog.jpg")
A = np.mean(A, axis=2)
plt.imshow(A)
plt.set_cmap('gray')
```

```python
U, S, VT = np.linalg.svd(A, full_matrices=1)

error_list = np.array([])
for r in np.arange(0,2000,10):
    UUstar = U[:r, :] @ U.T[:, :r]
    error = UUstar - np.eye((UUstar.shape[1]))
    error_list = np.append(error_list,[r,np.linalg.norm(error)])


# Reshape the error_list for plotting
error_list = error_list.reshape(-1, 2)
r_values = error_list[:, 0]  # Extract r values
error_norms = error_list[:, 1]  # Extract error norms

plt.scatter(r_values, error_norms)
    
```
